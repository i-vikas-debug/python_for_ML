#  creating imbalanced dataset and then doing upsampling

```python
import numpy as np
import pandas as pd
```

```python
n_samples = 1000
n_ratio = 0.9
n_class_0 = int(n_samples*n_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
class_0 = pd.DataFrame({'Feature_1': np.random.normal(loc=0,scale=1,size=900),'Target': [0]*n_class_0})     # normally distributed dataset
```

```python
class_0.head()
```

```python
class_1 = pd.DataFrame({'Feature_1': np.random.normal(loc=0,scale=1,size=100),'Target': [1]*n_class_1})     # normally distributed dataset
```

```python
class_1.head()
```

```python
df = pd.concat([class_0,class_1])
```

```python
df.head()
```

```python
df.tail()
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df[['Target']].value_counts()
```

### Here we will upsample where target =1 

```python
from sklearn.utils import resample
```

```python
df_major = df[df['Target']==0]
```

```python
df_minority = df[df['Target']==1]
```

```python
df_major.head()
```

```python
df_minority.head()
```

```python
df_minor_upsample = resample(df_minority,replace=True, n_samples = len(df_major),random_state=42)
```

```python
df_minor_upsample.shape
```

```python
df_upsampled = pd.concat([df_minor_upsample,df_major])
```

```python
df_upsampled['Target'].value_counts()
```

```python
sns.histplot(df_upsampled['Feature_1'],kde=True)
```

# Handling Outlier

```python
import numpy as np
import seaborn as sns
```

```python
lst_marks = [30,40,50,60,70,80,90,40,45,59,78,89,1000,1100]
```

```python
sns.boxplot(lst_marks )
```

### To handle outliers we will decide the lower , higher fence

```python
data_ka_25,data_50,data_75 = np.percentile(lst_marks,[25,50,75])
```

```python
data_ka_25,data_50,data_75
```

```python
min , Q1 , Q2 ,Q3 , max = np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
min, Q1 ,Q2 , Q3 , max
```

```python
IQR = Q3 - Q1
```

```python
IQR
```

```python
lower_fence = Q1-1.5* IQR
higher_fence = Q3 +1.5* IQR
```

```python
lower_fence,higher_fence
```
